```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
img = plt.imread("images.jpg")
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.subplot(1,2,2)
plt.imshow(gray, cmap='gray')
```

```python
def equalization(img):

     grayDistributed = np.zeros(256)
     for i in range(img.shape[0]):
          for  j in range(img.shape[1]):
               grayDistributed[int(img[i][j])] += 1

     equalizationProcess = np.zeros(256)
     equalizationProcess = grayDistributed / (img.shape[0] * img.shape[1])
     equalizationProcess = equalizationProcess * (256-1)
     equalizationProcess = np.cumsum(equalizationProcess)
     equalizationProcess = np.round(equalizationProcess)

     result = np.zeros(img.shape)
     for i in range(img.shape[0]):
          for  j in range(img.shape[1]):
               result[i][j] = equalizationProcess[int(img[i][j])]

     equaDistributed = np.zeros(256)
     for i in range(img.shape[0]):
          for  j in range(img.shape[1]):
               equaDistributed[int(result[i][j])] += 1
               
     return equaDistributed, result
               
```

```python
equahist, equa_img = equalization(gray)

hist = np.zeros(256)
for i in range(gray.shape[0]):
     for  j in range(gray.shape[1]):
          hist[int(gray[i][j])] += 1

grayindex = [i for i in range(256)]

plt.figure(figsize=(10, 10))
plt.subplot(2,2,1)
plt.imshow(gray, cmap='gray')
plt.subplot(2,2,2)
plt.imshow(equa_img, cmap='gray')

plt.subplot(2,2,3)
plt.bar(grayindex,hist)
plt.subplot(2,2,4)
plt.bar(grayindex,equahist)
```

```python
def specification(img, imgTarget):
     _, equaImage = equalization(img)
     _, equaTarget = equalization(imgTarget)

     result = np.zeros(img.shape)
     
     return result
```
